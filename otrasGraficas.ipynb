```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 16, 25]

plt.scatter(x, y, label='x^2', color='b', marker='o')
plt.title('Scatter Plot')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 16, 25]

plt.plot(x, y, label='x^2', color='b', marker='o', linestyle='-')
plt.title('Smooth Line Plot')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

categorias = ['A', 'B', 'C']
values = [7, 3, 5]

plt.pie(values, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Esto asegura que el grfico de pastel se vea circular.

plt.title('Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

categorias = ['A', 'B', 'C']
values1 = [7, 3, 5]
values2 = [4, 6, 2]

width = 0.35  # Ancho de las barras

fig, ax = plt.subplots()

ax.bar(categorias, values1, width, label='Valores 1')
ax.bar(categorias, values2, width, label='Valores 2', bottom=values1)

ax.set_xlabel('Categoras')
ax.set_ylabel('Valores')
ax.set_title('Grfico de Barras Apiladas')
ax.legend()

plt.show()


```
